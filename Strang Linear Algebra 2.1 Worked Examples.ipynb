## 2.1A
**Question:** Describe the column picture of these three equations *Ax* = *b*. Solve by careful inspection of the columns (instead of elimination).  

*x* + 3*y* + 2*z* = -3  
2*x* + 2*y* + 2*z* = -2  
3*x* + 5*y* + 6*z* = -5

This system of equations is fairly easy to solve because *b* is the second column multiplied by -1. Thus the veector *x* is (0,-1,0). We can prove this by looking at their intersections in 3D space. An equation with 3 unknowns is a plane. To get the planes I rearranged each equation to isolate *z*.

(*x* + 3*y* + 3)/-2 = z  
(*x* + 2*y* - 5)/3 = z  
(3*x* + 5*y* + 5)/6 = z

```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# this is an ugly way to set up the planes but it works
x_ax = list(range(-101, 101, 1))
y_ax = list(range(-101, 101, 1))

# prepare equation 1
x1, y1 = np.meshgrid(x_ax, y_ax)
z1 = (vx + 3*vy + 3)/-2

# prepare equation 2
x2, y2 = np.meshgrid(x_ax, y_ax)
z2 = (vx + 2*vy - 5)/3

# prepare equation 3
x3, y3 = np.meshgrid(x_ax, y_ax)
z3 = (3*vx + 5*vy + 5)/6

# plot the planes
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(x1, y1, z1, color='b', alpha=0.3, label='col. 1')
plt3d.plot_surface(x2, y2, z2, color='g', alpha=0.3, label='col. 1')
plt3d.plot_surface(x3, y3, z3, color='r', alpha=0.3, label='col. 1')
plt3d.scatter3D(0,-1,0, label='solution')

plt.show()
```

## 2.1B
**Question:** This system has no solution. The planes in the row picture don't meet at a point. No combination of the three columns produces b. How to show this?  
*x* + 3*y* + 5*z* = 4  
*x* + 2*y* - 3*z* = 5  
2*x* + 5*y* + 2*z* = 8


I can visualize this exactly as I did above, I just need to rearrange the equations.  
(*x* + 3*y* - 4)/-5 = z  
(*x* + 2*y* - 5)/3 = z  
(2*x* + 5*y* - 8)/-2 = z

```python
x_ax = list(range(-101, 101, 1))
y_ax = list(range(-101, 101, 1))

# prepare equation 1
x1, y1 = np.meshgrid(x_ax, y_ax)
z1 = (vx + 3*vy - 4)/-5

# prepare equation 2
x2, y2 = np.meshgrid(x_ax, y_ax)
z2 = (vx + 2*vy - 5)/3

# prepare equation 3
x3, y3 = np.meshgrid(x_ax, y_ax)
z3 = (2*vx + 5*vy - 8)/-2

# plot the planes
plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(x1, y1, z1, color='b', alpha=0.3, label='col. 1')
plt3d.plot_surface(x2, y2, z2, color='g', alpha=0.3, label='col. 1')
plt3d.plot_surface(x3, y3, z3, color='r', alpha=0.3, label='col. 1')
plt3d.scatter3D(4,5,8, label='solution')

plt.show()
```
