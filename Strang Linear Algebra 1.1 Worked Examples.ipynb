## 1.1A
**Question:** The linear combinations of *v* = (1, 1, 0) and *w* = (0, 1, 1) fill a plane in 3D space. *Describe that plane*. Find a vector that is not a combination of *v* and *w* - not on the plane.

```python
#import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

First let's describe *v* and *w*. They are both lines.

```python
xs = [0, 1, 100]
ys = [0, 1, 100]
zs = [0, 1, 100]
flat = [0, 0, 0]

plt3d = plt.figure().gca(projection='3d')
plt3d.plot(xs, ys, flat, label='v')
plt3d.plot(flat, ys, zs, label='w')
plt3d.legend()

plt.show()
```

Because *c* and *d* are all real numbers, *cv* (red) and *dw* (blue)are the planes of the xy axes and yz axes, respectively. No legend for this figure because of a bug in Matplotlib https://stackoverflow.com/questions/54994600/pyplot-legend-poly3dcollection-object-has-no-attribute-edgecolors2d

```python
x_ax = list(range(0, 101, 1))
y_ax = list(range(0, 101, 1))
z_ax = list(range(0, 101, 1))

vx, vy = np.meshgrid(x_ax, y_ax)
vz = vx*0 + vy*0

wy, wz = np.meshgrid(y_ax, z_ax)
wx = wy*0 + wz*0

plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(vx, vy, vz, color='r', alpha=0.5, label='cv')
plt3d.plot_surface(wx, wy, wz, color='b', alpha=0.5, label='dw')

#plt3d.legend()
plt.show()
```

*cv* + *dw* is the combination of both planes.

```python
c = list(range(0,101))

cv = []
dw = []

for i in c:
    cv.append([i,i,0])
    dw.append([0,i,i])

xs = []
ys = []
zs = []
for i in range(0, len(cv)):
    x = cv[i][0] + dw[i][0]
    y = cv[i][1] + dw[i][1]
    z = cv[i][2] + dw[i][2]
    xs.append(x)
    ys.append(y)
    zs.append(z)

plt3d = plt.figure().gca(projection='3d')
plt3d.scatter3D(xs, ys, zs, color='g')


plt.show()
```

As discussed in the book, any point on the resulting plane must be equal to (*c*, *c*+*d*, *d*). So any point where *x* + *z* ≠ *y* is not on the plane.

```python
c = list(range(0,101))

cv = []
dw = []

for i in c:
    cv.append([i,i,0])
    dw.append([0,i,i])

xs = []
ys = []
zs = []
for i in range(0, len(cv)):
    x = cv[i][0] + dw[i][0]
    y = cv[i][1] + dw[i][1]
    z = cv[i][2] + dw[i][2]
    xs.append(x)
    ys.append(y)
    zs.append(z)
    
out_xs = [20, 90]
out_ys = [10, 90]
out_zs = [80, 50]

plt3d = plt.figure().gca(projection='3d')
plt3d.scatter3D(xs, ys, zs, color='k')
plt3d.scatter3D(out_xs, out_ys, out_zs, color='r')


plt.show()
```

## 1.1B
**Question:** For *v* = (1,0) and *w* = (0,1), describe all points *cv* with (**1**) *whole numbers c* (**2**) *nonnegative* numbers *c* ≥ 0. Then add all vectors *dw* abd describe all *cv* + *dw*.

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# for convenience I'm restricting the visualization from -5 to 5
xmin = -5
xmax = 6
```

**Step 1** All whole numbers *c*. This is all integers from *-∞* to *∞*. *c1* in legend.

I used this to set up my number line. https://stackoverflow.com/questions/23186804/graph-point-on-straight-line-number-line-in-python

```python
int_c = list(range(xmin, xmax, 1))
y = [0] * len(int_c)

for x in int_c:
    plt.annotate(x,
                 (x,0),
                 textcoords='offset points',
                 xytext=(0,-15),
                 ha='center')
    
plt.vlines(0,0,1)
plt.plot(int_c,y,'o',color='r',label='c1')

plt.hlines(0,-6,6,color='k',linestyle='dotted',label='x-axis')
plt.vlines(0,0,1,color='k',label='0')
plt.xlim(-5.5,5.5)
plt.ylim(-1, 10)
plt.axis('off')
plt.legend()
```

**Step 2** *Nonnegative* numbers *c* ≥ 0. This is the right half of the number line. *c2* in legend.

```python
pos_c = list(range(0, xmax, 1))
y = [0] * len(pos_c)

for x in int_c:
    plt.annotate(x,
                 (x,0),
                 textcoords='offset points',
                 xytext=(0,-15),
                 ha='center')
    
plt.plot(pos_c, y,'-',color='r',label='c2')
plt.hlines(0,-6,6,color='k',linestyle='dotted',label='x-axis')
plt.vlines(0,0,1,color='k',label='0')
plt.xlim(-5.5,5.5)
plt.ylim(-1, 10)
plt.axis('off')
plt.legend()
```

**Step 3** All points in *c*. This is the combination of *c1* and *c2*, i.e. every integer from 0 to *∞*. *c* in legend. Because *v* = (1,0), this also describes *cv*.

```python
pos_int_c = [x for x in int_c if x >= 0]
y = [0] * len(pos_int_c)

for x in int_c:
    plt.annotate(x,
                 (x,0),
                 textcoords='offset points',
                 xytext=(0,-15),
                 ha='center')

plt.plot(pos_int_c,y,'o',color='r',label='c')
plt.hlines(0,-6,6,color='k',linestyle='dotted',label='x-axis')
plt.vlines(0,0,1,color='k',label='0')
plt.xlim(-5.5,5.5)
plt.ylim(-1, 10)
plt.axis('off')
plt.legend()
```

**Step 4**. Add all vectors *dw* and describe *cv* + *dw*. Because *w* = (0,1), it will extend a vertical line at every point in *cv*. 

```python
legend_flag = False
for x in pos_int_c:
    if legend_flag == False:
        plt.axvline(x,color='b',label='d')
        legend_flag = True
    else:
        plt.axvline(x,color='b')
        
for x in int_c:
    plt.annotate(x,
                 (x,0),
                 textcoords='offset points',
                 xytext=(0,-15),
                 ha='left')
    
plt.plot(pos_int_c,y,'o',color='r',label='c')    
plt.hlines(0,-6,6,color='k',linestyle='dotted',label='x-axis')
plt.vlines(0,0,1,color='k',label='0')
plt.xlim(-5.5,5.5)
plt.ylim(-1, 10)
plt.axis('off')
plt.legend()
```

## 1.1C
**Question:** Find two equations for *c* and *d* so that the linear combination of *cv* + *dw* = *b*.  

*v* = (2,-1)  
*w* = (1,-2)  
*b* = (1,0)

This is a simultaneous system of equations we can solve by elimination. I wrote a function a while ago that does this in an aesthetic (at least to me) way. I included a subset in the following code block so I could show it off.
2*c* - *d* = 1  
-*c* + 2*d* = 0

```python
SIG_DIGITS = 2

def findVal(x1, x2):
    v = float(x2 / x1)
    return v

def findY(y1, y2, z1, z2):
    y_f = (z1 - z2)/(y1 - y2)
    return y_f

def findX(x1, y1, y_f, z1):
    sub_y = y1 * y_f
    z1 = z1 - sub_y
    x_f = z1/x1
    return x_f

def multiplyEq(x, y, z, c):
    x *= c
    y *= c
    z *= c
    return x, y, z

def printEq(x, y, z, eq_num):
    print(str(round(x, SIG_DIGITS)) + "*x + " + str(round(y, SIG_DIGITS)) + "*y = " + str(round(z, SIG_DIGITS)) + "\t\t#" + str(eq_num))
    
def confirmValidPair(x1, y1, z1, x2, y2, z2):
    if (z1 != z2):
        if (x1 == x2) and (y1 == y2):
            return False
        else:
            return True
    else:
        return True

def stylizedElimination(x1, y1, z1, x2, y2, z2):
    printEq(x1, y1, z1, 1)
    printEq(x2, y2, z2, 2)
    if not (confirmValidPair(x1, y1, z1, x2, y2, z2)):
        print(confirmValidPair(x1, y1, z1, x2, y2, z2))
        print("this is not a valid simultaneous system of equations")
        return 1
    print("\n")
    v = findVal(x1, x2)
    if (x2 % x1 == 0):
        print("Multiply both sides of Eq. 1 by " + str(v))
    else:
        print("Multiply both sides of Eq. 1 by " + str(x2) + "/" + str(x1))
    x3, y3, z3 = multiplyEq(x1, y1, z1, v)
    printEq(x3, y3, z3, 3)
    print("\n")

    print("Substract Eq. 3 from Eq. 2")
    y_f = findY(y2, y3, z2, z3)
    print("Find that y = " + str(round(y_f, SIG_DIGITS)))

    print("Substitute y = " + str(round(y_f, SIG_DIGITS)) + " in Eq. 1")
    x_f = findX(x1, y1, y_f, z1)

    if (abs((x1*x_f + y1*y_f) - z1) < SIG_DIGITS):
        print("Answer\t x = " + str(round(x_f, SIG_DIGITS)) + "\t y = " + str(round(y_f, SIG_DIGITS)))
    return 0

stylizedElimination(2,-1,1,-1,2,0)
```

We can see that *c* should equal 2/3 and *d* should equal 1/3.

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

c = np.linspace(0, 3)
y1 = (2*c)-1
y2 = (c)/2

plt.plot(c,y1, label='2c-d=1')
plt.plot(c,y2, label='-c+2d=0')
plt.plot(0.67, 0.33, 'o',label='intersection')

plt.legend()
```
